# Titanic Dataset Overview

## The Titanic dataset is a well-known dataset that provides information on the passengers aboard the RMS Titanic, which sank after colliding with an iceberg in 1912. The dataset is commonly used for data analysis and machine learning practice.
### Columns Description
- PassengerId: Unique identifier for each passenger.
- Survived: Survival indicator (0 = No, 1 = Yes).
- Pclass: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
- Name: Name of the passenger.
- Sex: Gender of the passenger.
- Age: Age of the passenger.
- SibSp: Number of siblings/spouses aboard the Titanic.
- Parch: Number of parents/children aboard the Titanic.
- Ticket: Ticket number.
- Fare: Passenger fare.
- Cabin: Cabin number.
- Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).
### Summary of the Dataset
- The dataset contains 891 entries.
- The dataset includes both numeric and categorical data.
- Some columns have missing values:
- Cabin: 687 missing values.
- Embarked: 2 missing values.
### Goals of the Exploratory Data Analysis (EDA)
- Inspect the structure of the dataset: Understand the data types and identify missing values.
- Data Cleaning: Handle missing values appropriately.
- Explore Data Characteristics: Analyze distributions, relationships, and patterns within the data.
- Visualize Insights: Use histograms, scatter plots, heatmaps, and other visualizations to gain insights into the data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
# Load the Titanic dataset from the uploaded CSV file
titanic = pd.read_csv(r"C:\Users\mdsha\Downloads\titanic\train.csv")
```

```python
# Fill missing values in 'Cabin' with a placeholder
titanic['Cabin'].fillna('Unknown', inplace=True)

# Fill missing values in 'Embarked' with the most frequent value
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
# Plot distributions of numerical features
titanic.hist(bins=20, figsize=(14, 10))
plt.show()
```

```python
# Select only numeric columns
numeric_columns = titanic.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Plot correlation heatmap for numeric columns
plt.figure(figsize=(10, 8))
sns.heatmap(titanic[numeric_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot survival rate by sex
sns.barplot(x='Sex', y='Survived', data=titanic)
plt.show()
```

```python
# Plot survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=titanic)
plt.show()
```

```python
# Plot age distribution by survival
sns.kdeplot(data=titanic, x='Age', hue='Survived', fill=True)
plt.show()
```

```python
# Survival rate by port of embarkation
sns.barplot(x='Embarked', y='Survived', data=titanic)
plt.title('Survival Rate by Port of Embarkation')
plt.show()
```

```python
# Survival rate by number of siblings/spouses aboard
sns.barplot(x='SibSp', y='Survived', data=titanic)
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.show()
```

```python
# Survival rate by number of parents/children aboard
sns.barplot(x='Parch', y='Survived', data=titanic)
plt.title('Survival Rate by Number of Parents/Children Aboard')
plt.show()
```

```python
# Pair plot to examine relationships between numerical features
sns.pairplot(titanic, hue='Survived', vars=['Age', 'Fare', 'Pclass', 'SibSp', 'Parch'])
plt.show()
```

```python
# Survival rate by fare
sns.kdeplot(data=titanic, x='Fare', hue='Survived', fill=True)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
# Box plot for age by survival status
sns.boxplot(x='Survived', y='Age', data=titanic)
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
# Box plot for fare by survival status
sns.boxplot(x='Survived', y='Fare', data=titanic)
plt.title('Fare Distribution by Survival Status')
plt.show()
```

```python
# Violin plot for age by passenger class and survival status
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=titanic, split=True)
plt.title('Age Distribution by Passenger Class and Survival Status')
plt.show()
```

```python
# Violin plot for fare by passenger class and survival status
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=titanic, split=True)
plt.title('Fare Distribution by Passenger Class and Survival Status')
plt.show()
```

# Conclusion
## The Exploratory Data Analysis (EDA) of the Titanic dataset provided several key insights:

- Gender and Class: Women and children, as well as passengers in the 1st class, had higher chances of survival.
- Fare and Embarkation: Higher fares were associated with higher survival rates, and passengers who embarked from Cherbourg had a better chance of survival.
